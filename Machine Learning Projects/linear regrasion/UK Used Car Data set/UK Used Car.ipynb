# Preprocessing

```python

from sklearn.preprocessing import LabelEncoder
```

```python
df.head()
```

```python
# convert categorical values to numerical
df = df.replace({'transmission': {'Manual': 1, 'Automatic': 2, 'Semi-Auto':3}})
```

```python
df['fuelType'].unique()
```

```python
# convert categorical values to numerical
df = df.replace({'fuelType': {'Petrol': 1, 'Diesel': 2, 'Hybrid':3}})
```

```python
#convert categorical values to numerical using Laber Encoder
le = LabelEncoder()
df['model']= le.fit_transform(df['model'])

```

```python
df['model'].value_counts()
```

# Model Training

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
```

```python
model = RandomForestRegressor()

model.fit(x_train,y_train)
```

```python
pred = model.predict(x_train)
```

```python
error_score = r2_score(y_train, pred)
print('R squared Error: ', error_score * 100)
```

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
scores = cross_val_score(model, x, y, cv=cv, scoring='r2')
```

```python
print('Cross-Validation Scores:', scores)
print('Mean R-squared:', scores.mean())
print('Standard Deviation:', scores.std())
```
