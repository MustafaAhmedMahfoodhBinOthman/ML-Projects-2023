```python
# Calculate z-scores for absences and identify outliers
from scipy.stats import zscore

df1['absences_zscore'] = zscore(df1['absences'])
outliers = df1[(df1['absences_zscore'] > 3) | (df1['absences_zscore'] < -3)]
print('Outliers:', outliers)
```

# Data Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df1.columns
```

```python
training_x = ['sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout',
       'health','total_alcohol', 'absences_zscore', 'G1', 'G2',]
x = df1[training_x]
y = df1['G3']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
training_data_prediction = model.predict(x_train)
```

```python
error_score = r2_score(y_train, training_data_prediction)
print('R squared Error: ', error_score * 100)
```

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel('Actual G3')
plt.ylabel('Predicted G3')
plt.title('Actual G3 vs Predicted G3')
plt.show()
```

```python
test_data_prediction = model.predict(x_test)
```

```python
error_score = r2_score(y_test, test_data_prediction)
print('R squared Error: ', error_score * 100)
```

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel('Actual G3')
plt.ylabel('Predicted G3')
plt.title('Actual G3 vs Predicted G3')
plt.show()
```
